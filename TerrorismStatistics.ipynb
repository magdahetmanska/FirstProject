## Import data


```python
# import data from csv file to data farme
import pandas as pd
import numpy as np

data = pd.read_csv('globalterrorismdb_0617dist.csv', encoding='ISO-8859-1', low_memory=False)
# visualisation of global terrorism database, it isn't necessary, only to know how the data looks like (waste memory and time ;])
data
```

## Simple histogram

```python
import matplotlib.pyplot as plt

#Show two simple histograms 

year=data['iyear']
plt.hist(year,50,normed=0,alpha=0.75)

plt.title('Terrorist attacks in years');
plt.xlabel('Year');
plt.ylabel('Attack namber');
plt.grid(True);
plt.show()

month=data['imonth']
plt.hist(month,5,normed=0,alpha=0.5,color='g')

plt.title('Terrorist attacks in months');
plt.xlabel('Month');
plt.ylabel('Attack namber');
plt.grid(True);
plt.show()


```

## Plots

```python
plt.rcParams["figure.figsize"] = [10,5] #change size of figure

#This bar plot show in which region of the world attacks are the are the biggest danger
pd.value_counts(data['region_txt']).plot.bar()
plt.title('Amount of terrorist attacks in regions');
plt.xlabel('Region');
plt.ylabel('Attack namber');
plt.grid(True);
plt.show()

#This bar plot show in which region of the world attacks are the are the biggest danger
pd.value_counts(data['dbsource']).plot.bar()
plt.title('Origin of data');
plt.xlabel('Database');
plt.ylabel('Amount of data');
plt.grid(True);

plt.show()

#This line plots show how number of terrorist attack increase over the years
pd.value_counts(data['iyear'])
plt.title('Increase in attacks over the years');
plt.xlabel('Year');
plt.ylabel('Attack namber');
plt.grid(True);
plt.plot(list(set(data["iyear"])),(data.groupby('iyear')).count())
plt.show()
```
